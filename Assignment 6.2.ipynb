```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head()
```

```python
titanic.describe()
```

```python
fig, ax = plt.subplots()
titanic["sex"].value_counts().plot.pie(autopct='%1.0f%%')
ax.set_title("Number Of Passengers By Sex")
plt.show()
```

```python
colors = ['b', 'r']

m = titanic[titanic.sex=='male']
f = titanic[titanic.sex=='female']

plt1 = plt.scatter(m["age"], m["fare"], marker='o', color=colors[0])
plt2 = plt.scatter(f["age"], f["fare"], marker='o', color=colors[1])

plt.legend((plt1, plt2),
           ('male', 'female'),
           loc='upper right')

plt.show()
```
